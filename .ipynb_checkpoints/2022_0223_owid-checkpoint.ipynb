```python
url="C:/Users/phat/OneDrive/Documents/country_code.xlsx"
country_code=pd.read_excel(url)
country_code
```

```python
iso_emu = country_code.alpha3[country_code.emu == 1].tolist()
iso_emu
```

```python
data_owid_emu = data_owid[data_owid.iso_code.isin(iso_emu)]
data_owid_emu
```

```python
emu_newcases7davg_country = data_owid_emu.pivot(index='date', columns='iso_code', values = 'new_cases_smoothed')
emu_newcases7davg = emu_newcases7davg_country.sum(axis = 1, skipna = True)
print(emu_newcases7davg)
emu_vac_country = data_owid_emu.pivot(index='date', columns='iso_code', values = 'total_vaccinations').fillna(method='ffill')
emu_vac = emu_vac_country.sum(axis = 1, skipna = True)
print(emu_vac)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
emu_newcases7davg.plot()
plt.show()
emu_vac.plot()
plt.show()
```
